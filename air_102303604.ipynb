```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import zipfile
import os

# ---------------- STEP 1: Extract ZIP ----------------
zip_path = "/content/archive (2).zip"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("/content/")

# Check extracted files
print(os.listdir("/content"))

# ---------------- STEP 2: Load Data ----------------
data = pd.read_csv("/content/data.csv", encoding="latin1")

print(data.head())
print(data.columns.tolist())

# ---------------- STEP 3: Extract NO2 column ----------------
no2 = data["no2"].dropna()
print("Total NO2 samples:", len(no2))

# ---------------- STEP 4: Define roll number constants ----------------
# Roll number = 102303604
# r mod 7 = 4  -> ar = 0.20
# r mod 5 = 4  -> br = 0.7

ar = 0.20
br = 0.7

# ---------------- STEP 5: Apply transformation ----------------
z = no2 + ar * np.sin(br * no2)

# ---------------- STEP 6: Estimate mean (μ) ----------------
mu = np.mean(z)
print("Mean (mu) =", mu)

# ---------------- STEP 7: Estimate standard deviation (σ) ----------------
sigma = np.std(z)
print("Standard deviation (sigma) =", sigma)

# ---------------- STEP 8: Calculate lambda (λ) ----------------
lam = 1 / (2 * sigma**2)
print("Lambda =", lam)

# ---------------- STEP 9: Calculate constant c ----------------
c = 1 / (np.sqrt(2*np.pi) * sigma)
print("c =", c)

# ---------------- STEP 10: Plot histogram ----------------
plt.hist(z, bins=40, density=True)
plt.title("Histogram of transformed variable z")
plt.xlabel("z values")
plt.ylabel("Density")
plt.show()

# ---------------- STEP 11: Plot PDF ----------------
x_vals = np.linspace(min(z), max(z), 500)
pdf = c * np.exp(-lam * (x_vals - mu)**2)

plt.hist(z, bins=40, density=True)
plt.plot(x_vals, pdf)
plt.title("Histogram vs Learned PDF")
plt.xlabel("z values")
plt.ylabel("Density")
plt.show()
```
